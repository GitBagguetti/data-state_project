```python
# Install nltk if not present
# !pip install nltk

import nltk
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download tokenizer data (NLTK 3.9+ needs punkt_tab for sentence tokenization)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

# Concatenate all comments into a single string, dropping NaN values
comments_text = ' '.join(arms_df['Comments'].dropna().astype(str))

# Tokenize the words using English language specialization
tokens = word_tokenize(comments_text.lower(), language='english')

# Remove common English stopwords and non-alphabetic tokens
# stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.isalpha()]

# Count the most frequent words
word_freq = Counter(filtered_tokens)
```

```python
# Show the most common words
print("Most frequent words in 'Comments':")
for word, freq in word_freq.most_common(40):
    print(f"{word}: {freq}")
```

```python
# Define the size of n-grams to extract
N = 2  # Change this value to get different n-grams (e.g., N=2 for bigrams, N=3 for trigrams)

# Download tokenizer data (NLTK 3.9+ needs punkt_tab for sentence tokenization)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

# Concatenate all comments into a single string, dropping NaN values
comments_text = ' '.join(arms_df['Comments'].dropna().astype(str))

# Tokenize the words using English language specialization
tokens = word_tokenize(comments_text.lower(), language='english')

# Remove common English stopwords and non-alphabetic tokens
# stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.isalpha()]

# Generate n-grams from filtered tokens
def ngrams(tokens, n):
    return zip(*[tokens[i:] for i in range(n)])

ngrams_list = [' '.join(gram) for gram in ngrams(filtered_tokens, N)]

# Count the most frequent n-grams
ngram_freq = Counter(ngrams_list)
```

```python
# Show the most common words
print(f"Most frequent {N}_grams in 'Comments':")
for word, freq in ngram_freq.most_common(40):
    print(f"{word}: {freq}")
```
